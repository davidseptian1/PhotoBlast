```python
import os
import smtplib
from email.mime.text import MIMEText
from email.mime.image import MIMEImage
from email.mime.multipart import MIMEMultipart
import time
import tqdm
import pandas as pd

def kirim_email_dengan_gambar_attachment(email_pengirim, email_password, penerima, template, subject, photo_dir):
    """
    Mengirim email dengan template HTML dan attachment foto dari direktori
    
    Parameters:
    - email_pengirim: email pengirim
    - email_password: password email pengirim
    - penerima: email penerima
    - template: path ke file template HTML
    - subject: subjek email
    - photo_dir: direktori yang berisi foto-foto
    """
    # Konfigurasi email
    msg = MIMEMultipart('mixed')
    msg['Subject'] = subject
    msg['From'] = email_pengirim
    msg['To'] = penerima

    # Membaca dan menambahkan template HTML
    with open(template, 'r') as f:
        html = f.read()
    html_part = MIMEText(html, 'html')
    msg.attach(html_part)

    # Menambahkan foto-foto sebagai attachment
    photo_files = [f for f in os.listdir(photo_dir) if f.startswith('photo_') and f.endswith('.png')]
    photo_files.sort()  # Mengurutkan file agar terurut (photo_0.png, photo_1.png, dst)

    for photo_file in photo_files:
        file_path = os.path.join(photo_dir, photo_file)
        with open(file_path, 'rb') as f:
            img = MIMEImage(f.read())
            img.add_header('Content-Disposition', 'attachment', filename=photo_file)
            msg.attach(img)

    try:
        # Mengirim email
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(email_pengirim, email_password)
            smtp.send_message(msg)
            time.sleep(0.15)
        print(f"Email berhasil dikirim ke {penerima} dengan {len(photo_files)} attachment foto")
    except Exception as e:
        print(f"Gagal mengirim email ke {penerima}: {e}")
```

```python
import os
import smtplib
from email.mime.text import MIMEText
from email.mime.image import MIMEImage
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart
from email import encoders
import time
import tqdm
import pandas as pd

def attach_file(msg, file_path):
    """
    Helper function to attach any file to email message
    """
    with open(file_path, 'rb') as f:
        # For images
        if file_path.lower().endswith(('.png', '.jpg', '.jpeg')):
            attachment = MIMEImage(f.read())
        # For other files (including videos)
        else:
            attachment = MIMEBase('application', 'octet-stream')
            attachment.set_payload(f.read())
            encoders.encode_base64(attachment)
        
        # Add header
        filename = os.path.basename(file_path)
        attachment.add_header('Content-Disposition', 'attachment', filename=filename)
        msg.attach(attachment)

def kirim_email_dengan_media(email_pengirim, email_password, penerima, template, subject, photo_dir, video_path):
    """
    Mengirim email dengan template HTML, attachment foto dari direktori, dan video
    
    Parameters:
    - email_pengirim: email pengirim
    - email_password: password email pengirim
    - penerima: email penerima
    - template: path ke file template HTML
    - subject: subjek email
    - photo_dir: direktori yang berisi foto-foto
    - video_path: path ke file video
    """
    # Konfigurasi email
    msg = MIMEMultipart('mixed')
    msg['Subject'] = subject
    msg['From'] = email_pengirim
    msg['To'] = penerima

    # Membaca dan menambahkan template HTML
    with open(template, 'r') as f:
        html = f.read()
    html_part = MIMEText(html, 'html')
    msg.attach(html_part)

    # Menambahkan foto-foto sebagai attachment
    photo_files = [f for f in os.listdir(photo_dir) if f.startswith('photo_') and f.endswith('.png')]
    photo_files.sort()  # Mengurutkan file

    for photo_file in photo_files:
        file_path = os.path.join(photo_dir, photo_file)
        attach_file(msg, file_path)

    # Menambahkan video sebagai attachment
    if os.path.exists(video_path):
        attach_file(msg, video_path)

    try:
        # Mengirim email
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(email_pengirim, email_password)
            smtp.send_message(msg)
            time.sleep(0.15)
        print(f"Email berhasil dikirim ke {penerima} dengan {len(photo_files)} foto dan video")
    except Exception as e:
        print(f"Gagal mengirim email ke {penerima}: {e}")

# def kirim_email_batch(email_pengirim, email_password, data_file, template, subject, photo_dir, video_path):
#     """
#     Mengirim email ke banyak penerima dari file Excel/CSV
    
#     Parameters:
#     - email_pengirim: email pengirim
#     - email_password: password email pengirim
#     - data_file: path ke file Excel/CSV yang berisi daftar email penerima
#     - template: path ke file template HTML
#     - subject: subjek email
#     - photo_dir: direktori yang berisi foto-foto
#     - video_path: path ke file video
#     """
#     # Baca file data penerima
#     if data_file.endswith('.csv'):
#         df = pd.read_csv(data_file)
#     else:
#         df = pd.read_excel(data_file)
    
#     # Kirim email ke setiap penerima
#     for index, row in tqdm.tqdm(df.iterrows(), total=len(df)):
#         penerima = row['email']  # Sesuaikan dengan nama kolom email di file
#         kirim_email_dengan_media(
#             email_pengirim,
#             email_password,
#             penerima,
#             template,
#             subject,
#             photo_dir,
#             video_path
#         )
#         time.sleep(1)  # Delay antar pengiriman

# Contoh penggunaan:
# email_pengirim = "your_email@gmail.com"
# email_password = "your_app_password"  # Gunakan App Password jika 2FA aktif
# penerima = "recipient@email.com"
# template = "template.html"
# subject = "PhotoBlast - Your Photos and Video"
# photo_dir = "photo"
# video_path = "output_slideshow.mp4"  # Path ke file video

# # Untuk satu penerima
# kirim_email_dengan_media(
#     email_pengirim,
#     email_password,
#     penerima,
#     template,
#     subject,
#     photo_dir,
#     video_path
# )

# Untuk banyak penerima
# data_file = "penerima.xlsx"
# kirim_email_batch(email_pengirim, email_password, data_file, template, subject, photo_dir, video_path)
```
